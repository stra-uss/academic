```python
import gradio as gr
from pyecharts.charts import Bar
from pyecharts import options as opts

def show_chart():
    bar = (Bar().add_xaxis(["A", "B"]).add_yaxis("Count", [1, 2]))
    # Use render_embed to get the html
    html = bar.render_embed()
    return f'<iframe style="width: 100%; height: 400px" srcdoc="{html}"></iframe>'

demo = gr.Interface(fn=show_chart, inputs=[], outputs="html")
demo.launch()
```

```python
import gradio as gr
import plotly.express as px
import pandas as pd
import numpy as np

# 1. Função que gera o gráfico Plotly
def create_plot(df_size, plot_type):
    # Gerar dados aleatórios
    df = pd.DataFrame({
        "x": np.random.randn(df_size),
        "y": np.random.randn(df_size),
        "category": np.random.choice(["A", "B", "C"], df_size)
    })

    # Criar gráfico interativo
    if plot_type == "Dispersão":
        fig = px.scatter(df, x="x", y="y", color="category", title="Gráfico de Dispersão")
    else:
        fig = px.box(df, y="x", title="Boxplot")

    return fig

# 2. Configurando a Interface Gradio
with gr.Blocks() as demo:
    gr.Markdown("# Gradio + Plotly Deploy")

    with gr.Row():
        slider = gr.Slider(10, 100, value=50, step=1, label="Tamanho dos Dados")
        dropdown = gr.Dropdown(["Dispersão", "Boxplot"], value="Dispersão", label="Tipo de Gráfico")

    plot_output = gr.Plot() # Componente para Plotly
    button = gr.Button("Gerar Gráfico")

    # Conectar entrada e saída
    button.click(create_plot, inputs=[slider, dropdown], outputs=plot_output)

# 3. Lançar
demo.launch()
```
